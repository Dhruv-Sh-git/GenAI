```python
from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage, SystemMessage
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
import os

# Set your API key
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

model = ChatGoogleGenerativeAI(
    model="models/gemini-pro",
    streaming=False  # Disabled to reduce token usage on free tier
)
model
```

```python
### create message
### system instruction is the prompt given to the model to set behavior
messages = [
    SystemMessage(content="You are a helpful assistant"),
    HumanMessage(content="what are the benefits of using langchain?")
]
```

```python
response=model.invoke(messages)
response
```

```python
print(response.content)
```

```python
# streaming example
for chunk in model.stream(messages):
    print(chunk.content, end='', flush=True)
```

```python
# dynamic prompt template
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate
## create translation app
translation_template = ChatPromptTemplate.from_messages(  [
    SystemMessagePromptTemplate.from_template("You are a helpful assistant that translates {input_language} to {output_language}."),
    HumanMessagePromptTemplate.from_template("Translate the following text: {text}")  
])
prompt=translation_template.invoke({
    "input_language":"English",
    "output_language":"French",
    "text":"I love programming."
    })
```

```python
translated_response= model.invoke(prompt)
print(translated_response.content)
```
